```python
import tensorflow as tf 
from tensorflow import keras 
```

```python
import matplotlib.pyplot as plt 
import numpy as np
from keras.utils import to_categorical
import random
```

```python
mnist = tf.keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data() 
```

```python
train_images = train_images / 255 
test_images = test_images / 255 
```

```python
train_labels = to_categorical(train_labels, 10) 
test_labels = to_categorical(test_labels, 10) 
```

```python
df1 = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(256, activation='softmax'),
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation='softmax'),
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
df1.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
df1.fit(train_images, train_labels, batch_size=128, epochs=20)
test_loss, test_acc = df1.evaluate(test_images, test_labels, verbose=0)
```

```python
n = random.randint(0, 9999) 
plt.imshow(test_images[n]) 
plt.show() 
prediction = df1.predict(test_images) 
print("Prediction %d" % np.argmax(prediction[n])) 
```

```python
test_loss, test_acc = df1.evaluate(test_images, test_labels) 
print("Loss = %.3f" % test_loss) 
print("Accuracy = %.3f" % test_acc)
```

```python
df2 = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation='softmax'),
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
df2.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])
df2.fit(train_images, train_labels, batch_size=128, epochs=20)
test_loss, test_acc = df2.evaluate(test_images, test_labels, verbose=0)
```

```python
df3 = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
df3.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
df3.fit(train_images, train_labels, batch_size=128, epochs=20)
test_loss, test_acc = df3.evaluate(test_images, test_labels, verbose=0)
```
